```python
import json
from dotenv import load_dotenv, find_dotenv
from pathlib import Path
import os
from pandas import json_normalize

from football_modelling.plotting import Field, add_possession_chain_sb
```

```python
match_id_required = 69301
home_team_required = "England Women's"
away_team_required = "Sweden Women's"
```

```python
load_dotenv(find_dotenv(), override=True)
data_root_dir = Path(os.path.expanduser(os.getenv("DATA_ROOT_DIR")))
statsbomb_dir = data_root_dir.joinpath(os.getenv("STATSBOMB_DIR"))
events_dir = statsbomb_dir.joinpath("events")
file_name = str(match_id_required) + ".json"
file_path = events_dir.joinpath(file_name)
```

```python
with open(file_path) as file:
    data = json.load(file)
```

```python
match_events = json_normalize(data, sep="_")
```

```python
relevant_possessions = match_events[(match_events["type_id"] == 16)][
    "possession"
].unique()
```

```python
relevant_possessions
```

```python
ground = Field()
ground.add_title("Possession chains")
```

```python
for possession in relevant_possessions:
    df = match_events[match_events["possession"] == possession]

    add_possession_chain_sb(ground.figure, df, team_name=df.iloc[0]["team_name"])

ground.save("Output/possessions.html")
```
